```python
import json
# 由于文件中有多行，直接读取会出现错误，因此一行一行读取
file = open("./work_dirs/resnet101_paris_zurich/20210724_112551.log.json", 'r', encoding='utf-8')
papers = []
for line in file.readlines():
    dic = json.loads(line)
    papers.append(dic)


print(len(papers))
```

```python
iteration = []
loss = []
acc = []
for i in range(1,len(papers)):
    if papers[i]['mode'] == 'train':
        iteration.append(papers[i]['iter'])
        loss.append(papers[i]['decode.loss_seg'])
        acc.append(papers[i]['decode.acc_seg'])
```

```python
# Train Berlin 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(iteration[:1170],loss[:1170])
plt.title("resnet101-paris-zurich", fontsize=20) 
plt.xlabel("iteration", fontsize=15) 
plt.ylabel("loss_value", fontsize=15) 
plt.show() #顯示繪製的圖形
```

```python
plt.plot(iteration,acc)
plt.title("Accuracy", fontsize=20) #圖表標題
plt.xlabel("iteration", fontsize=15) #x軸標題
plt.ylabel("pixel_acc", fontsize=15) #y軸標題
plt.show() #顯示繪製的圖形
```

```python
import json
# 由于文件中有多行，直接读取会出现错误，因此一行一行读取
file = open("./work_dirs/swin-small_potsdam_berlin/20210718_075001.log.json", 'r', encoding='utf-8')
papers = []
for line in file.readlines():
    dic = json.loads(line)
    papers.append(dic)

papers = papers[:-203]
count = 0
file = open("./work_dirs/swin-small_potsdam_berlin/20210719_085032.log.json", 'r', encoding='utf-8')
for line in file.readlines():
    if count == 0:
        count = 1
        continue
    else:
        dic = json.loads(line)
        papers.append(dic)
print(len(papers))
iteration = []
loss = []
acc = []
for i in range(1,len(papers)):
    if papers[i]['mode'] == 'train':
        iteration.append(papers[i]['iter'])
        loss.append(papers[i]['decode.loss_seg'])
        acc.append(papers[i]['decode.acc_seg'])
```
